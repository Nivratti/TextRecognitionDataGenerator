```python
# Yield successive n-sized
# chunks from l.
def divide_chunks(l, n):
      
    # looping till length l
    for i in range(0, len(l), n): 
        yield l[i:i + n]
  
# # How many elements each
# # list should have
# n = 3000000
  
# lst_sublists = list(divide_chunks(list(set_all_newspaper_words), n))
```

```python
import os, io
from nb_utils.file_dir_handling import list_files
from tqdm import tqdm 

dir_chunks_all_newspaper_unique_words = "/content/drive/MyDrive/deep_learning/ocr_arabic/dataset/text-corpus/arabic_billion_words/chunks-all-newspaper-unique-words"

out_dir_smaller_chunks = "smaller-chunks-500k-each-min-char-2-max-30"
os.makedirs(out_dir_smaller_chunks, exist_ok=True)

txt_files = list_files(dir_chunks_all_newspaper_unique_words, filter_ext=[".txt"], maxdepth=1)

i = 0
large_words = []
for txt_file in tqdm(txt_files):
    lst_single_newspaper_words = []
    with io.open(txt_file, "r", encoding="utf-8") as f:
        # print(f.readlines()[:2])
        for word in f.readlines():
            if word:
                if len(word) <= 30:
                    lst_single_newspaper_words.append(word.strip("\n"))
                else:
                    # print(f"word: ", word)
                    large_words.append(word)
              
    # How many elements each
    # list should have
    n = 500000
      
    lst_sublists = list(divide_chunks(list(lst_single_newspaper_words), n))

    for sublist in lst_sublists:
      i += 1
      out_filename = os.path.join(out_dir_smaller_chunks, f"chunks-{i:02}-cnt-{len(sublist)}.txt")
      with io.open(out_filename, "w", encoding="utf-8") as f:
          f.write("\n".join(sublist))
```
