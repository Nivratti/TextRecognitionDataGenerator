```python
lst_raw_txt = dataset["train"]["text"][0].split(" ")
print(lst_raw_txt)
```

```python
import regex

for raw_txt in lst_raw_txt:
    only_arabic = regex.sub(r'[^\u0600-\u06FF]', u'', raw_txt)
    print(only_arabic)
```

```python
from tqdm import tqdm

all_arabic_processed_words = []
for single_text_str in tqdm(dataset["train"]["text"]):
    # print(single_text_str)
    lst_raw_txt = single_text_str.split(" ")

    for raw_txt in lst_raw_txt:
        only_arabic = regex.sub(r'[^\u0600-\u06FF]', u'', raw_txt)
        # print(only_arabic)

        if only_arabic and (len(only_arabic) > 2):
          all_arabic_processed_words.append(only_arabic)
    # break
```

```python
print(len(all_arabic_processed_words))
```

```python
print(len(all_arabic_processed_words))
```

```python
## Remove duplicates
# using set()
# to remove duplicated 
# from list 
lst_unique_all_arabic_processed_words = list(set(all_arabic_processed_words))
```

```python
len(lst_unique_all_arabic_processed_words)
```

```python
# using list comprehension to
# perform removal
test_list = [i for i in lst_unique_all_arabic_processed_words if i]
```

```python
test_list
```

```python
import io

with io.open(r"Sabanews-unique-words.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(lst_unique_all_arabic_processed_words[:10]))
```
